### Customer Profiling based on Calculation of Probabilities - Conditional, Marginal Probability

#### Marginal Probabilities
The Marginal Probabilities are shown at the margins of the Contingency tables. It is basically the "Unconditional Probability" and simply refers that an event is not conditioned on another event.

```python
# Calculating Marginal Probabilities using Contingency Table containing Gender and Product
pd.crosstab(index = df['Gender'], columns = df['Product'], normalize = True, margins = True)
```

Here the probabilities occuring at the margins of the "Contingency Table" are the "Marginal Probabilities". So, the marginal probability of Product type "KP281" is 0.444, for "KP481" is 0.333 and for "KP781" is 0.222. Whereas the marginal probability of Females is 0.422 while that of Males is 0.5778.

```python
# Calculating Marginal Probabilities using Contingency Table containing Marital Status and Product
pd.crosstab(index = df['MaritalStatus'], columns = df['Product'], normalize = True, margins = True)
```

The marginal probabilities for different types of product has already been shown above. But for the "MaritalStatus", the marginal probabilities are 0.4055 for "Single" customers while 0.5944 for "Partnered" customers. 

```python
# Calculating Marginal Probabilities using Contingency Table containing Usage and Product
pd.crosstab(index = df['Usage'], columns = df['Product'], normalize = True, margins = True)
```

```python
# Calculating Marginal Probabilities using Contingency Table containing Fitness and Product
pd.crosstab(index = df['Fitness'], columns = df['Product'], normalize = True, margins = True)
```

#### Conditional Probabilities
Conditional Probability is the probability of occurence any event A, given that an other Event B has already occured. So it basically tells us that what is the effect of Event B on happening of Event A if Event B has already occured.  

```python
# Contingency Table showing Gender and Product columns
pd.crosstab(index = df['Gender'], columns = df['Product'], margins = True)
```

```python
# P[A|B] = P[A intersection B]/P[B]
# Eg: prob_using_KP281_given_female = probability_of_female_using_KP281/probability_of_KP281 
prob_using_KP281_given_female = 40/80
prob_using_KP281_given_male = 40/80
print(f"P[KP281 | Female]: {prob_using_KP281_given_female}")
print(f"P[KP281 | Male]: {prob_using_KP281_given_male}")
prob_using_KP481_given_female = 29/60
prob_using_KP481_given_male = 31/60
print(f"P[KP481 | Female]: {prob_using_KP481_given_female}")
print(f"P[KP481 | Male]: {prob_using_KP481_given_male}")
prob_using_KP781_given_female = 7/40
prob_using_KP781_given_male = 33/40
print(f"P[KP781 | Female]: {prob_using_KP781_given_female}")
print(f"P[KP781 | Male]: {prob_using_KP781_given_male}")
```

From the above calculation of conditional probabilities, we can say that:

(1) Probability of using KP281 given that the customer is a female is 0.5

(2) Probability of using KP481 given that the customer is a female is 0.483

(3) Probability of using KP781 given that the customer is a female is 0.175

(4) Probability of using KP281 given that the customer is a male is 0.5

(5) Probability of using KP481 given that the customer is a male is 0.516

(6) Probability of using KP781 given that the customer is a male is 0.825

```python
# Contingency Table showing Marital Status and Product columns 
pd.crosstab(index = df['MaritalStatus'], columns = df['Product'], margins = True)
```

```python
# P[A|B] = P[A intersection B]/P[B]
# Eg: prob_using_KP281_given_partnered = probability_of_partnered_using_KP281/probability_of_KP281 
prob_using_KP281_given_partnered = 48/80
prob_using_KP281_given_single = 32/80
print(f"P[KP281 | Partnered]: {prob_using_KP281_given_partnered}")
print(f"P[KP281 | Single]: {prob_using_KP281_given_single}")
prob_using_KP481_given_partnered = 36/60
prob_using_KP481_given_single = 24/60
print(f"P[KP481 | Partnered]: {prob_using_KP481_given_partnered}")
print(f"P[KP281 | Single]: {prob_using_KP481_given_single}")
prob_using_KP781_given_partnered = 23/40
prob_using_KP781_given_single = 17/40
print(f"P[KP781 | Partnered]: {prob_using_KP781_given_partnered}")
print(f"P[KP781 | Single]: {prob_using_KP781_given_single}")
```

From the above calculation of conditional probabilities, we can say that:

(1) Probability of using KP281 given that the customer is Partnered is 0.6

(2) Probability of using KP481 given that the customer is Partnered is 0.6

(3) Probability of using KP781 given that the customer is Partnered is 0.575

(4) Probability of using KP281 given that the customer is Single is 0.4

(5) Probability of using KP481 given that the customer is Single is 0.4
